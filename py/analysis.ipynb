```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from typing import List
```

```python
colors = ["blue", "red", "green", "orange", "purple", "teal"]

def make_plot(title: str, names: List[str]):
    for i, nm in enumerate(names):
        df = pd.read_csv(f"../data/{nm}.csv")
        plt.plot(df["n"], df["time"], color=colors[i], label=nm)
        plt.fill_between(df["n"], df["time"] - df["stddev"], df["time"] + df["stddev"], color=colors[i], alpha=.1)

    plt.title(title)
    plt.rcParams["figure.figsize"] = (16, 9)
    plt.legend()
    plt.xscale("log")
    plt.yscale("log")
    plt.xlabel("Size")
    plt.ylabel("Time (ms)")
    plt.savefig("../tmp/plot.png")
    plt.show()
```

```python
make_plot("Vector Addition",
          ["vecadd_forloop_cpu-py",
           "vecadd_numpy_cpu-py",
           "vecadd_torch_cpu-py",
           "vecadd_torch_gpu-py",
           "vecaddForloopCpu-js"])
           
```

```python
make_plot("Square Matrix Multiplication",
          ["matmul_forloop_cpu-py",
           "matmul_numpy_cpu-py",
           "matmul_torch_cpu-py",
           "matmul_torch_gpu-py",
          ])
```
